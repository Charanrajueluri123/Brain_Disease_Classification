```python
import os
import pydicom
from pydicom.dataset import Dataset
from pydicom.dataset import FileDataset
from pydicom.uid import ExplicitVRLittleEndian  #  Add this for Transfer Syntax UID
from PIL import Image
import numpy as np

# Define input and output folder
input_folder = ""
output_folder = ""
os.makedirs(output_folder, exist_ok=True)

def convert_jpg_to_dcm(jpg_path, dcm_path):
    # Open image and convert to grayscale
    img = Image.open(jpg_path).convert("L")
    np_img = np.array(img)

    # Create File Meta Information
    file_meta = pydicom.dataset.FileMetaDataset()
    file_meta.MediaStorageSOPClassUID = pydicom.uid.SecondaryCaptureImageStorage
    file_meta.MediaStorageSOPInstanceUID = pydicom.uid.generate_uid()
    file_meta.TransferSyntaxUID = ExplicitVRLittleEndian  #  Add this line

    # Create a DICOM dataset
    ds = FileDataset(dcm_path, {}, file_meta=file_meta, preamble=b"\0" * 128)
    ds.Modality = "OT"  # Other modality
    ds.SOPClassUID = pydicom.uid.SecondaryCaptureImageStorage
    ds.Rows, ds.Columns = np_img.shape
    ds.PhotometricInterpretation = "MONOCHROME2"
    ds.SamplesPerPixel = 1
    ds.BitsStored = 8
    ds.BitsAllocated = 8
    ds.HighBit = 7
    ds.PixelRepresentation = 0
    ds.PixelData = np_img.tobytes()

    # Save as DICOM
    ds.save_as(dcm_path, write_like_original=False)  #  Ensure full DICOM writing

# Process all JPG files
for filename in os.listdir(input_folder):
    if filename.lower().endswith(".jpg"):
        jpg_path = os.path.join(input_folder, filename)
        dcm_path = os.path.join(output_folder, filename.replace(".jpg", ".dcm"))
        convert_jpg_to_dcm(jpg_path, dcm_path)
        print(f"Converted: {filename} -> {dcm_path}")

print("Conversion complete!")
```
